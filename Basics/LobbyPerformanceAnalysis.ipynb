```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
perSec=pd.read_csv("runtimes_every1.txt", header=None, names = ["Time (ms)"])
perSec = perSec.div(1000000)
plt.figure()
perSec.plot()
avg1 = perSec.mean()
std1 = perSec.std()[0]
plt.plot([0,100],[avg1,avg1], label = 'Average = %f ms' % (avg1))
plt.legend()
plt.xlabel('client number')
plt.ylabel('time (ms)')
plt.title('100 clients in 100 seconds')
plt.savefig('100sec.png')
```
